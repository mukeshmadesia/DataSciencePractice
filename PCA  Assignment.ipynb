```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
wine=pd.read_csv("D:\\ExcelR Assignments\\PCA Assignment\\wine.csv")
print(wine.describe())
wine.head()
```

```python
wine1=wine.iloc[:,1:]
wine1.head()
```

```python
WINE=wine1.values
WINE
```

```python
wine_normal=scale(WINE)
wine_normal
```

```python
pca=PCA()
pca_values=pca.fit_transform(wine_normal)
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
pca.components_
```

```python
plt.plot(var1,color='black')
```

```python
final_df=pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
fig=plt.figure(figsize=(6,12))
sns.scatterplot(data=final_df)
```

# Checking with Hierarchial Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
wine.head()
```

```python
def minmax_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
wine_norm=minmax_func(wine.iloc[:,1:])
wine_norm
```

```python
plt.figure(figsize=(12,10))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(wine_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters.value_counts()
```

```python
y_hc
```

```python
wine1['h clusterid']=y_hc
wine1
```

# Checking with KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
wine.head()
```

```python
def minmaxfunc(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
def norm_func(i):
    x=(i-i.mean())/(i.std())
    return(x)
```

```python
x=norm_func(wine.iloc[:,1:])
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_wine_df=scaler.fit_transform(wine.iloc[:,1:])
scaled_wine_df
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_wine_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
cluster = KMeans(3, random_state=42)
cluster.fit(scaled_wine_df)
```

```python
cluster.labels_
```

```python
wine['clusterid_new']=cluster.labels_
wine
```
